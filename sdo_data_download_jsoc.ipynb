```python
'''
SDO data search and download code through JSOC  
'''

import drms
import astropy.time
from sunpy.net import Fido, attrs as a
from sunpy.net import jsoc


client = jsoc.JSOCClient()

print(a.jsoc.Series)

res = client.search(a.Time('2023-10-02T12:00:00', '2023-10-02T12:00:15'),
                    a.jsoc.Series('aia.lev1_euv_12s'),
                    a.jsoc.PrimeKey('WAVELNTH', '304'),
                    a.jsoc.Notify('qingfeng_pang@foxmail.com'))  


client = drms.Client()
print(client.pkeys('aia.lev1_euv_12s'))

res
```

```python
instrument_keyword = res[0]['INSTRUME']
download_files_start_time = res[0]['T_REC']
print(instrument_keyword)

download_dir = f'/data2/pqf_SDO_SolO/data_SDO/{instrument_keyword}_start_from_{download_files_start_time}'

download_res = Fido.fetch(res, path=download_dir)
```
